- Adicionando uma coluna nova no Pandas: pode ser ap artir de uma ja existente, ou voce criar uma nova com valor padrão

```python
# Criar um coluna 'comissão' com base nas vendas
Vendas_df['Comissão'] = Vendas_df['Valor Final'] *0.05
display(Vendas_df)

# Criar uma coluna com valor padrão
Vendas_df.loc[:, "Imposto"] = 0   # Na coluna de imposto eu vou preencher todas as linhas com zero
display(Vendas_df)

#  Os ' : ' significa 'todos'
```

```python
# Adicionar linhas na tabela, repare que nossas vendas vão até novembro somrente
# então vamos adicionar as vendas de dezembro

# Vizualizar a tabela de dezembro
Vendas_dez_df = pd.read_excel("Vendas - Dez.xlsx")
# display(Vendas_dez_df)
if 'Vendas_df' not in globals():
    Vendas_df = pd.DataFrame()

Vendas_df = pd.concat([Vendas_df, Vendas_dez_df], ignore_index=True)  # metodo 'append' para adicionar também
display(Vendas_df)

# ignore_index=True é utilizado no método concat do pandas para ignorar os índices originais
# dos DataFrames sendo concatenados e gerar novos índices sequenciais para o DataFrame resultante.
```

- Calcular Indicadores: Groupby, Value Counts.

```python
# Calcular as transações que cada loja fez? 
Transacoes_Loja = Vendas_df['ID Loja'].value_counts() # Somar os valores
display(Transacoes_Loja)
```

```python
# Groupby (vai agrupar as informações)
# Eu quero saber o faturamento dos produtos
# Exemplo, vou agrupar um produto e somar todos os valores desse produto

# Fatura_Produtos = Vendas_df.groupby('Produto').sum()
# display(Fatura_Produtos)

# Selecionar somente algumas colunas [Produto, Valorfinal]
Fatura_Produtos = Vendas_df[['Produto', 'Valor Final']].groupby('Produto').sum()
display(Fatura_Produtos)
```
