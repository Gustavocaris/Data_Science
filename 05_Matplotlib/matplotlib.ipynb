### Gráficos Python - Matplotlib.pyplot

- Matplotlib é uma biblioteca em Python amplamente utilizada para visualização de dados e criação de gráficos. Ele oferece uma ampla variedade de funcionalidades para criar gráficos estáticos, interativos e animados em diferentes formatos e estilos. Ele pode gerar gráficos em vários formatos de saída, incluindo PNG, PDF, SVG e muitos outros. Isso torna fácil incorporar os gráficos em documentos, apresentações ou páginas da web.




- biblioteca: https://matplotlib.org/stable/tutorials/introductory/pyplot.html


```python
import matplotlib.pyplot as plt
```

### plot()


- é sempre bom ir lendo a documentação, tem bastante coisa nova aqui 
- Mas o "plot" é p/ referir a qualquer gráfico ou figura que você cria usando a biblioteca

```python
# Lembra do plano cartesiano, eixo X e Y
x = [10, 20, 30, 40] 
y = [20, 30, 40, 30]

plt.plot(x, y)
plt.show()
```

#### aqui da uma olhada nas propriedades da plot, tem a imagem daquilo que é editavel dentro do nosso grafico

```python
plt.plot(x, y, label='dados')
plt.ylabel('Eixo Y')
plt.xlabel('Eixo X')
plt.title('Gustavo aula')
plt.xticks([0,10,20,30,40,50,60,70,80])
plt.yticks([0,20,40,60,80,100])
plt.legend()
plt.show()
```

### Lines propriedades

### .Scatter()

- é um tipo de gráfico usado para visualizar a relação entre duas variáveis. Ele exibe pontos no plano cartesiano, onde cada ponto representa uma observação nos dados. Um eixo geralmente representa uma variável e o outro eixo representa a outra variável. O scatter plot é útil para identificar padrões, tendências ou relações entre as variáveis.


- documentação: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html#matplotlib-pyplot-scatter

```python
plt.scatter(x, y)
plt.show()
# A GENTE Ta olhando os mesmos dados de antes... apenas explorando esses graficos;
```

```python

# Fixing random state for reproducibility
np.random.seed(19680801)

# Compute areas and colors
N = 150
r = 2 * np.random.rand(N)
theta = 2 * np.pi * np.random.rand(N)
area = 200 * r**2
colors = theta

fig = plt.figure()
ax = fig.add_subplot(projection='polar')
c = ax.scatter(theta, r, c=colors, s=area, cmap='hsv', alpha=0.75)
```

### .bar()

docs: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html#matplotlib-pyplot-bar

### Alterando os limites dos gráficos:
doc> https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.axis.html#matplotlib.pyplot.axis

```python
plt.plot(x,y)
plt.axis(xmin=-1,xmax=20,ymin=-5,ymax=15)
plt.axis('auto')
plt.axis('square')
plt.show()
```
