```python
import pandas as pd
import plotly.express as px
# importando os modulos 

tabela = pd.read_csv('cancelamentos_sample.csv')
# ler os arquivos e armazenar em um DataFrame do pandas chamado tabela
# DataFrame é uma estrutura de dados tabular bidimensional em Pandas, semelhante a uma planilha do Excel.

coluna_escolhida = 'idade'

grafico = px.histogram(tabela, x=coluna_escolhida, color="cancelou")
grafico.show()
```

```python
tabela = pd.read_csv('cancelamentos_sample.csv')

# -> Definindo faixas etárias
faixa_etaria = pd.cut(tabela['idade'], bins=[0, 18, 30, 40, 50, 60, 70, float('inf')], labels=['0-18', '19-30', '31-40', '41-50', '51-60', '61-70', '71+'])
#Nesta linha, estamos usando a função pd.cut() do pandas para dividir a coluna 'idade' do DataFrame tabela em faixas etárias.
# As faixas etárias são definidas pelos argumentos bins, que são os limites das faixas, e labels,
# que são os rótulos para as faixas. As faixas de idade são de 0 a 18, 19 a 30, 31 a 40, e assim por diante.

# -> Criando um novo DataFrame com as faixas etárias
dados_faixa_etaria = pd.DataFrame({'Faixa Etária': faixa_etaria})

# Contando o número de pessoas em cada faixa etária
contagem_faixa_etaria = dados_faixa_etaria['Faixa Etária'].value_counts().reset_index()

# Renomeando as colunas
contagem_faixa_etaria.columns = ['Faixa Etária', 'Número de Pessoas']

# Criando o gráfico de pizza
grafico = px.pie(contagem_faixa_etaria, values='Número de Pessoas', names='Faixa Etária', title='Distribuição de Idade')
grafico.show()
```
