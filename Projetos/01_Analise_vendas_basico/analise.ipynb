```python
# Primeiro passo : Trazer a base de dados e ver o que tem nela
import pandas as pd

tabela = pd.read_excel("vendas.xlsx")
display(tabela)
```

```python
# dei uma olhada no geral da tabela, pensa assim, eu quero aumentar o meu faturamento, certo? 
# então vamo dar uma olhada no faturamento geral então, depois a gente vai penerando cada vez mais

faturamento_total = tabela["Valor Final"].sum() #.sum é pra somar a coluna que eu selecionei
print(faturamento_total)
```

```python
# voce tem que ir trabalhando um logica de raciocinio aqui, olha o valor total ali em cima
# e começa a pensar: de onde ele veio? afinal tem mais de uma filial... alguma vende + que a outra ??
# outra vende bem menos...

faturamento_por_loja = tabela[["ID Loja", "Valor Final"]]
display(faturamento_por_loja)
```

```python
# ele me deu o faturamento por loja agora, mas eu tenho que facilitar minha vida tambem,
# vou agrupar o nome das mesmas lojas somando os faturamentos. isso torna mais pratico e eficiente
# ----vou usar o 'groupby' e novemnte o 'sum'

faturamento_por_loja = tabela[["ID Loja", "Valor Final"]].groupby("ID Loja").sum()
display(faturamento_por_loja)
```

```python
# Olha ali, viu a diferença da iguatemi campinas para as demais filias ???? 
# essa é a sacada, começe pelo 'grosso' e vai filtrando aos poucos, agora é tentar descobrir o motivo dessa 
# diferença alarmante e tentar aumentar a venda das demais, vou pegar o produto agora também

faturamento_por_produto = faturamento_por_loja = tabela[["ID Loja", "Produto", "Valor Final"]].groupby(["ID Loja", "Produto"]).sum()
display(faturamento_por_produto)
```
