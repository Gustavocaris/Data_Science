```python
import pandas as pd

tabela = pd.read_csv("ClientesBanco.csv", encoding="latin1")
tabela = tabela.drop("CLIENTNUM", axis=1) # drop é para excluir um coluna, que ta dificultando um pouco, que nao serve
display(tabela)
# encoding-> é pra evitar erros de 'DeCode' que são devidos a alguns caracteres especificos por causa de idioma e tudo mais
```

```python
# tabela.info -> da algumas informações gerais sobre a tabela
display(tabela.info())
# reparando na 'categoria de cartão' ele tem apenas uma linha faltando,
# voce tambem pode preencher com alguma informação
```

```python
# vou excluir todas as linhas que tem apenas um valor vazio
tabela = tabela.dropna()
display(tabela.info())
# agora eu tenho 10126 valores não vazio

display(tabela.describe().round(2)) # '.round' é pra arredondar as casas numericas
# mostrar dados sob outros parametros, como os valores estão distribuidos conforme as categorias(colunas com numeros)
```
